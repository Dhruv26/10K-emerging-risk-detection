```python
import os
import glob
from tqdm import tqdm

from keybert import KeyBERT
from config import Config
```

```python
model = KeyBERT('distilbert-base-nli-mean-tokens')
```

```python
keywords = {}
for risk_file in tqdm(glob.glob(os.path.join(Config.risk_dir(), 'appl_risk', '*.txt'))):
    with open(risk_file) as f:
        doc = f.read()
    keywords[risk_file] = model.extract_keywords(doc, keyphrase_ngram_range=(1, 3), stop_words=None, top_n=25, diversity=0.2)
```
