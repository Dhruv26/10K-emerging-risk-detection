```python
import glob
import os

from rake_nltk import Rake
from tqdm import tqdm

from config import Config
```

```python
rake = Rake(max_length=3)
```

```python
keywords_1 = {}
for risk_file in tqdm(glob.glob(os.path.join(Config.risk_dir(), 'appl_risk', '*.txt'))):
    with open(risk_file) as f:
        doc = f.read()
    rake.extract_keywords_from_text(doc)
    keywords_1[risk_file] = rake.get_ranked_phrases()
```
