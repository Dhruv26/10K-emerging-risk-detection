```python
import glob
import os

from rake_nltk import Rake
from tqdm import tqdm

from config import Config
```

```python
rake = Rake()
```

```python
keywords = {}
for risk_file in tqdm(glob.glob(os.path.join(Config.risk_dir(), '*.txt'))):
    with open(risk_file) as f:
        doc = f.read()
    rake.extract_keywords_from_text(doc)
    keywords[risk_file] = rake.get_ranked_phrases_with_scores()
```
